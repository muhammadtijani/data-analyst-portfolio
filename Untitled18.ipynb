```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the orders
df = pd.read_csv('olist_orders_dataset.csv')

# Convert dates to real "Date" format
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])

print("Data Loaded!")
```

```python
import numpy as np

# 1. Calculate the 'Recency' (Days since the last order in the whole dataset)
snapshot_date = df['order_purchase_timestamp'].max()
df_recency = df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days
}).rename(columns={'order_purchase_timestamp': 'days_since_last_purchase'})

# 2. Create the Plot
plt.figure(figsize=(12, 6))
sns.histplot(df_recency['days_since_last_purchase'], bins=50, color='#2ecc71', kde=True)

# 3. Add labels that make it look professional
plt.title('Customer Churn Risk: Days Since Last Purchase', fontsize=15)
plt.xlabel('Days Since Last Order', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.axvline(x=180, color='red', linestyle='--', label='High Risk (6 Months)')
plt.legend()
plt.show()
```
